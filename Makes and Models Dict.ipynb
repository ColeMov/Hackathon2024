```python
import pandas as pd
cars = pd.read_excel("vehicle_data.xlsx")
```

```python
car_dict = {}

for index, row in cars.iterrows():
    make = row['make']
    year = row['year']
    model = row['model']

    if make not in car_dict:
        car_dict[make] = {}

    if year not in car_dict[make]:
        car_dict[make][year] = set()

    car_dict[make][year].add(model)

def sort_models(models):
    return sorted(models, key=lambda x: (not str(x).isdigit(), str(x).lower() if not str(x).isdigit() else ''))

for make, make_years in car_dict.items():
    for year, models_set in make_years.items():
        models_list = sort_models(models_set)
        car_dict[make][year] = models_list
```
